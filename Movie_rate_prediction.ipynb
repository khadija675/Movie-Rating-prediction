Classifying movies based on Ratings[Excellent,Good and Average]

```python
def rating_group(rating):
  if rating>=7.5:
    return 'Excellent'
  elif rating<7.5 and rating>=6:
    return 'Good'
  else:
    return 'Average'
```

```python
data['Rating_group']=data['Rating'].apply(rating_group)
```
